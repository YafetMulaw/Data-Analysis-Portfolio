```python
def male_female_child(passenger):
    age,sex = passenger

    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age', 'Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df.head()
```

```python
sns.countplot(x='Pclass', hue='person', data=titanic_df)
plt.show()
```

```python
fig = sns.FacetGrid(titanic_df, hue='Sex', aspect=4)
# `shade` is now deprecated in favor of `fill`; setting `fill=True`
fig.map(sns.kdeplot,'Age',fill=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
plt.show()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
```

```python
# FutureWarning: Passing `palette` without assigning `hue` is deprecated. Assign the `x` variable to `hue` and set `legend=False` for the same effect.
sns.countplot(x='Cabin', hue='Cabin', data=cabin_df, palette='summer')
plt.show()
```

```python
cabin_df = cabin_df[cabin_df.Cabin !='T']
sns.countplot(x='Cabin', hue='Cabin', data=cabin_df, palette='bone')
plt.show()
```

```python
sns.catplot(x='Cabin', hue='Cabin', data=cabin_df, palette='bone', kind='count')
plt.show()
```

```python
sns.countplot(x='Embarked', data=titanic_df, hue='Pclass', order=['C','Q','S'])
plt.show()
```

```python
sns.catplot(x='Embarked', data=titanic_df, hue='Pclass', order=['C','Q','S'], kind='count')
plt.show()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df.head()
```

```python
sns.catplot(x='Alone', data=titanic_df, hue='Alone', palette='Blues', kind='count', legend=False)
plt.show()
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})
```

```python
sns.catplot(x='Survivor', data=titanic_df, hue='Survivor', palette='Set1', legend=False, kind='count')
plt.show()
```

```python
# count', 'bar', 'box', 'violin', 'strip', 'swarm', 'point'
sns.catplot(x='Pclass', y='Survived', data=titanic_df, kind='point')
plt.show()
```

```python
sns.catplot(x='Pclass', y='Survived', hue='person', data=titanic_df, kind='point')
plt.show()
```
